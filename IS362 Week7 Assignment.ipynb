```python
import pandas as pd
```

```python
df = pd.read_csv('Week7HWratings.csv')
```

```python
df = df.rename(columns={"Unnamed: 0": "User"})
```

```python
df = df.dropna(subset=["User"])
```

```python
df = df.set_index("User")
```

```python
print(df)
```

Question 2: Show the average rating for each user and movie

```python
movie_cols = df.columns[1:]
```

Average rating by each user

```python
user_avg = df[movie_cols].mean(axis=1)
```

Average rating for each movie

```python
movie_avg = df[movie_cols].mean()
```

Display user average and movie average

```python
print("Average rating by user:")
print(user_avg)

print("\nAverage rating by movie:")
print(movie_avg)
```

Question 3: Create a dataframe with normalized rating

```python
normalized_df = df.copy()

normalized_df[movie_cols] = normalized_df[movie_cols].sub(user_avg, axis=0)

print(normalized_df)
```

Average normalized rating:

```python
norm_user_avg = normalized_df[movie_cols].mean(axis=1)
norm_movie_avg = normalized_df[movie_cols].mean()

print("Average normalized rating per user:")
print(norm_user_avg)

print("\nAverage normalized rating per movie:")
print(norm_movie_avg)
```

Normalized ratings adjust for how each person typically rates movies. Some users rate everything high, while others are more lower. By subtracting a user’s average rating, we can see whether they liked a movie more or less than they usually do.
It is beneficial and fair, as it lets us compare users whose rating habits are very different. Someone who always gives 5s and someone who mostly gives 3s become easier to compare once their ratings are normalized.
The downside is that normalized scores are less intuitive because they’re no longer on the familiar 1–5 scale. They can also be misleading when a user has rated only a few movies.
Overall, normalized ratings help recommendation systems focus on true preferences rather than raw scoring habits.
